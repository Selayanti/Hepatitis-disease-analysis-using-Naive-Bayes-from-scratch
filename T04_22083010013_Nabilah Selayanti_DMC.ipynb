```python
# Buat fungsi untuk menghitung mean
def mean(values):
    return sum(values) / float(len(values))

# Buat fungsi untuk menghitung standar deviasi
def stdev(values):
    avg = mean(values)
    variance = sum([(x - avg) ** 2 for x in values]) / float(len(values) - 1)
    return np.sqrt(variance)

# Buat fungsi untuk menghitung PDF (Probability Density Function)
def gaussian_probability(x, mean, stdev):
    exponent = np.exp(-((x - mean) ** 2 / (2 * stdev ** 2)))
    return (1 / (np.sqrt(2 * np.pi) * stdev)) * exponent

# Buat kelas Naive Bayes
class NaiveBayes:
    def __init__(self):
        self.summaries = {}

    # Training model
    def fit(self, X_train, y_train):
        self.classes = np.unique(y_train)
        for class_val in self.classes:
            self.summaries[class_val] = [(mean(attribute), stdev(attribute)) for attribute in zip(*X_train[y_train == class_val])]

    # Prediksi
    def predict(self, X_test):
        predictions = []
        for row in X_test:
            probabilities = {class_val: 1 for class_val in self.classes}
            for class_val in self.classes:
                for i in range(len(self.summaries[class_val])):
                    mean, stdev = self.summaries[class_val][i]
                    probabilities[class_val] *= gaussian_probability(row[i], mean, stdev)
            predictions.append(max(probabilities, key=probabilities.get))
        return predictions

# Bagi data menjadi fitur (X) dan variabel target (Y)
X = data.drop('Class', axis=1).values
Y = data['Class'].values

# Bagi data menjadi set pelatihan dan pengujian
def train_test_split(x,y):
    # split data train 70 % and test 30 %
    x_train, x_test, y_train, y_test = train_test_split(x, y,train_size=0.3,random_state=42)
    #normalization
    x_train_N = (x_train-x_train.mean())/(x_train.max()-x_train.min())
    x_test_N = (x_test-x_test.mean())/(x_test.max()-x_test.min())

# Model selection
naive_bayes_classifier = NaiveBayes()

# Model training
naive_bayes_classifier.fit(X_train, y_train)

# Melakukan prediksi pada set pengujian
predictions = naive_bayes_classifier.predict(X_test)


def classification_report(y_true, y_pred):
    unique_classes = np.unique(np.concatenate((y_true, y_pred)))
    report = ""
    for class_val in unique_classes:
        true_positive = np.sum((y_true == class_val) & (y_pred == class_val))
        false_positive = np.sum((y_true != class_val) & (y_pred == class_val))
        false_negative = np.sum((y_true == class_val) & (y_pred != class_val))
        precision = true_positive / float(true_positive + false_positive)
        recall = true_positive / float(true_positive + false_negative)
        f1_score = 2 * precision * recall / (precision + recall)
        report += f"Class {class_val}:\n"
        report += f"  Precision: {precision:.2f}\n"
        report += f"  Recall: {recall:.2f}\n"
        report += f"  F1 Score: {f1_score:.2f}\n"
    return report

# Menghitung akurasi
def accuracy_score(y_true, y_pred):
    correct = 0
    for i in range(len(y_true)):
        if y_true[i] == y_pred[i]:
            correct += 1
    return correct / float(len(y_true))

accuracy = accuracy_score(y_test, predictions)

# Mencetak hasil evaluasi
print("Accuracy:", accuracy)
print("-"*30)
print("Classification Report:")
print(report)
```
